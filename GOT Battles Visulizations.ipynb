```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline



battles = pd.read_csv('battles.csv')
```

```python
attacker = ('Joffrey/Tommen Baratheon', 'Robb Stark', 'Balon/Euron Greyjoy', 'Stannis Baratheon')
y_pos = np.arange(len(attacker))
performance = [14, 10, 7, 5]
 
plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, attacker)
plt.xlabel('# of Attacks')
plt.title('# of Times as Attacking General')
```

```python
labels = 'Joffrey/Tommen Baratheon', 'Robb Stark', 'Balon/Euron Greyjoy', 'Stannis Baratheon'
sizes = [14, 10, 7, 5]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

# 3. Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
battles.head()
## lets choose attacker king and attacker size
```

```python
sns.swarmplot(x="attacker_size", y="attacker_king", data=battles,
              size=5, palette="pastel", linewidth=1)
```

```python
sns.boxplot(x="attacker_size", y="attacker_king", data=battles,
            )
```

```python
sns.relplot(x="attacker_size", y="attacker_king", data=battles)
```

```python
sns.factorplot(x="attacker_size", y="attacker_king", data=battles, join=False)
```

```python
sns.relplot(x="attacker_size", y="attacker_king", 
            col="attacker_king", col_wrap=2,
           data=battles)
```

```python
sns.catplot(x="attacker_king", y="attacker_size", hue="defender_size",
            col="region", col_wrap=3, aspect=.6,
            kind="swarm", data=battles);
```

```python
sns.catplot(x="attacker_size", y="attacker_king", hue="major_death",
            kind="violin", split=True, data=battles);
```
